```python
import numpy as np

def getBondPriceV(y, face, couponRate, m, ppy=1):
   
    C = face * couponRate / ppy
    
    t = np.arange(1, m * ppy + 1)
    
    discount_factors = (1 + y / ppy) ** t
    
    pv_coupons = C / discount_factors
    
    pv_face = face / ((1 + y / ppy) ** (m * ppy))
    
    bond_price = pv_coupons.sum() + pv_face
    
    return bond_price
```

```python

y = 0.03
face = 2000000
couponRate = 0.04
m = 10
ppy = 2

print(getBondPriceV(y, face, couponRate, m, ppy))
```

```python
import numpy as np

def getBondDurationV(y, face, couponRate, m, ppy=1):
   
    C = face * couponRate / ppy

    t = np.arange(1, m * ppy + 1)

    CF = np.full_like(t, C, dtype=float)
    CF[-1] += face

    discount_factors = (1 + y / ppy) ** t

    PVCF = CF / discount_factors

    weighted_PVCF = t * PVCF

    duration = weighted_PVCF.sum() / PVCF.sum()

    return duration
```

```python
y = 0.03
face = 2000000
couponRate = 0.04
m = 10
ppy = 1

print("Duration =", round(getBondDurationV(y, face, couponRate, m, ppy), 2))
```

```python
import numpy as np

def FizzBuzz(start, finish):
    numvec = np.arange(start, finish + 1)
    objvec = np.array(numvec, dtype=object)

    objvec[(numvec % 3 == 0) & (numvec % 5 == 0)] = "fizzbuzz"
    objvec[(numvec % 3 == 0) & (numvec % 5 != 0)] = "fizz"
    objvec[(numvec % 5 == 0) & (numvec % 3 != 0)] = "buzz"

    return objvec
```

```python
print(FizzBuzz(61, 75))
```
